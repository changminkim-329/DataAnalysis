$${X_{new} = {{X - X_{mean}}\over{X_{std}}}}\$$

속성'BMI'가 정규화 되면 다음과 같이 된다.

```python
mean = data[selected_column].mean()
std = data[selected_column].std()

data[selected_column] = (data[selected_column]-mean)/std
```

## 박스 플롯
- **사분위**: 자료를 크기 순으로 배열하고, 누적 백분율을 4 등분한 각 점에 해당하는 값
    - 1사분위: 0~25%(Q1)
    - 2사분위: 0~50%(Q2)
    - 3사분위: 0~75%(Q3)
    - 4사분위: 0~100%(Q4)
- **중위값**: 변수을 크기 순서대로 늘어놓았을 때 그들의 한가운데 있는 값(Median)
<br>


- 박스는 25%(1사분위)에서 75%(3사분위)까지 값들을 박스로 둘러 싼다.
- 박스의 각 모서리 (Q1,Q3)로부터 IQR(Q3-Q1)의 1.5배 내에 값(whiskers:수염)이 있을 때 정상값으로 간주하고, 그렇지 않으면 이상치 값으로 간주한다.

### 밑의 함수는 자동으로 정상범위의 안에 있는 값만 추출해서 다시 박스플롯을 그리고  평균과 중앙값의 차이를 보여준다.

```python
def box_plot(data,selected_column):   
    q1 = data[selected_column].describe()['25%']
    q3 = data[selected_column].describe()['75%']
    iqr = (q3 - q1)

    plt.figure(figsize=(15,4))

    plt.subplot(1,2,1)
    plt.title(selected_column+' 데이터에서 이상치 삭제 전')
    plt.boxplot(data.dropna()[selected_column],vert=0)
    plt.axvline(np.mean(data.dropna()[selected_column]),color='r',label='평균값',linestyle='--')
    plt.legend()

    print('이상치 삭제전 |평균-중앙값|:',round(
          np.abs(np.mean(data.dropna()[selected_column])-np.median(data.dropna()[selected_column])),4))

    plt.subplot(1,2,2)
    plt.title(selected_column+' 데이터에서 이상치 삭제 후')


    left_whiskers = (data[selected_column]>=q1-1.5*iqr)
    right_whiskers = (data[selected_column]<=q3+1.5*iqr)
    datas = (data.dropna()[left_whiskers&right_whiskers])[selected_column]
    plt.boxplot(datas,vert=0)
    plt.axvline(np.mean(datas),color='r',linestyle='--',lw=2.0,label='평균값')
    plt.legend()
    print('이상치 삭제후 |평균-중앙값|:',round(np.abs(np.mean(datas)-np.median(datas)),4))
    plt.show()
    
    
    # return datas
```

```python
box_plot(data,selected_column)
```

```python
def multi_boxplot(data,hue=None):
    # 연속형 변수이름만 추출
    num_data = []
    for i in range(len(data.dtypes)):
        if(not((str(data.dtypes[i])=="category")or(str(data.dtypes[i])=="object"))):
            num_data.append(data.dtypes.index[i])
    plt.figure(figsize=(20,30))
    
    # 시각화
    if(hue != None):
        for i in range(len(num_data)):

            datas = data[[hue,num_data[i]]]
            q1 = datas[num_data[i]].describe()['25%']
            q3 = datas[num_data[i]].describe()['75%']
            iqr = (q3 - q1)

            plt.subplot(len(num_data)*2,2,2*i+1)
            plt.title(num_data[i]+' Before Clearing Outlier')
            sns.boxplot(x=hue,y=num_data[i],data=datas)
            for j in (datas[hue].unique()):

                plt.axhline(np.mean(datas[datas[hue]==j][num_data[i]]),color='r',
                            label=j+'_mean',linestyle='--')
                print('이상치 제거전 '+num_data[i]+'_'+j+ '|평균-중앙값|:',
                      round(np.abs(np.mean(datas[datas[hue]==j])
                             -np.median(datas[datas[hue]==j][num_data[i]]))[0],2))
            plt.legend()


            plt.subplot(len(num_data)*2,2,2*i+2)
            plt.title(num_data[i]+' After Clearing Outlier')
            left_whiskers = (datas[num_data[i]]>=q1-1.5*iqr)
            right_whiskers = (datas[num_data[i]]<=q3+1.5*iqr)
            datas = datas[left_whiskers&right_whiskers]
            sns.boxplot(x=hue,y=num_data[i],data=datas)
            for j in (datas[hue].unique()):
                plt.axhline(np.mean(datas[datas[hue]==j][num_data[i]])
                            ,color='r',linestyle='--',lw=2.0,label=j+'mean')
                print("이상치 제거후 ",num_data[i]+'_'+j+ '|평균-중앙값|:',
                      round(np.abs(np.mean(datas[datas[hue]==j][num_data[i]])
                          -np.median(datas[datas[hue]==j][num_data[i]])),2)
                   )
            plt.legend()
            print()
        plt.show()
        
    else:
        print("good")
```

### 원하지 않는 속성은 뺴고 시각화를 분석하는 bar_plot 함수

```python
def bar_plots(remove=None,hue=None):
    """
    remove = none,
    - 필요없는 분석은 뺴는 속성.(mean, std, median, max, min, size,'sum')
    - 리스트로 전달해야됨. ex) remove=['mean', 'median']
    
    hue = none,
    - 카테고리 변수 이름을 지정하여 케테고리 값에 따라 색상을 다르게 할 수 있다.
    - ex) hue="age"
    """
    np_math = {"mean":np.mean, "std":np.std, "sum":np.sum, "size":np.size,"median":np.median,
               "max":np.max, "min":np.min}
    if(remove!=None):
        for i in range(len(remove)):
            np_math.pop(remove[i])
    
    plt.figure(figsize=(20,5*(np.ceil(len(np_math)/3))))
    np_math_name_list = list(np_math.keys())
    for i in range(len(np_math)):
        plt.subplot(np.ceil(len(np_math)/3),3,i+1)
        plt.title(np_math_name_list[i])
        sns.barplot(x=selected_x_column,y=selected_y_column,data=data,
                    estimator=np_math[np_math_name_list[i]],hue=hue)

    plt.show()
```

### std, min, sum기능은 필요없어서 뺐다.

```python
sns.swarmplot(x="age",y="suicides_no",data=data,hue="sex")
```

## stripplot
- 스캐터 플롯처럼 도든 데이터를 점으로 표현하며, jitter=True를 설정하면 가로축상의 위치를 무작위로 바꾸어서 데이터의 수가 많을 경우에겹치지 않도록 한다.

```python
sns.stripplot(x="age",y="suicides_no",data=data,hue="sex",jitter=True)
```

## violinplot
- 세로 방향으로 커널 밀도 히스토그램을 그려주는데 왼쪽과 오른쪽이 대치잉 되도록 하여 바이올린처럼 표현.

```python
sns.violinplot(x="sex",y="suicides_no",data=data)
```

## pairplot
- 연속형 변수가 3개 이상일 경우, 그리드 형태로 각 변수의 조합에 대해 스캐터 플롯을 표현 하고 같은 변수가 만나는 대각션 영역에는 해당 데이터의 히스토그램을 그린다.

```python
selected_x_column = ['year']
selected_y_column = ['population','suicides_no_population']
hue = 'age'# 범주형 데이터
```
